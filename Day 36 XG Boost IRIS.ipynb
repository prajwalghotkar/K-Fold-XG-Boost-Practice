```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(iris.data,iris.target,test_size=0.2)
print(iris.data.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Create Train and Test Model - xgboost

''' Do you know ? prajwal 
    
    XGBoost comes with its own class for storing dataste called DMatrix. DMatrix is an internal data structure that is used
    by XGBoost,which is optimized for both memory efficiency and training speed'''
    
import xgboost as xgb
train=xgb.DMatrix(x_train,label=y_train) 
test=xgb.DMatrix(x_test,label=y_test)   
```

```python
print(train)
print(test)
```

```python
# Create Parameters

param={'max_depth':4,'eta':0.3,'objective':'multi:softmax','num_class':3}
epochs=10
```

```python
# Create Model

model=xgb.train(param,train,epochs)
display(model)
```

```python
# Prediction

prediction=model.predict(test)
print('prediction')
print(prediction)
print()
print('y test')
print(y_test)
```

```python
# Accuracy Score

from sklearn.metrics import accuracy_score
display(accuracy_score(y_test,prediction)) 
```

```python
# Create Decision Tree

from sklearn.tree import DecisionTreeClassifier
classifier=DecisionTreeClassifier(criterion='entropy')
classifier.fit(x_train,y_train)
print(classifier)
```

```python
# Predictions

y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
# Accuracy Score

print('Accuracy Score')
print(accuracy_score(y_test,y_pred),'\n')
```

```python
# Random Forest Classifier

from sklearn.ensemble import RandomForestClassifier
RF=RandomForestClassifier(n_estimators=100,criterion='entropy')
RF.fit(x_train,y_train)
print(RF) 
```

```python
# Prediction and Accuracy 
RF_y_pred=RF.predict(x_test)
RF_y_pred
print('Accuracy Score')
print(accuracy_score(y_test,RF_y_pred),'\n')
```
