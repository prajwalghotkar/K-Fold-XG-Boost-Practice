```python
# Data Set : Social Network Ads

# Import Libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
''' Read the Data'''

df1=pd.read_csv('Social_Network_Ads.csv')
print(df1)
```

```python
# CreateX Variable

x=df1.iloc[:,2:4].values
x
```

```python
# Create Y Variable

y=df1.iloc[:,4].values
print(y)
```

```python
# Feature Scaling

from sklearn.preprocessing import StandardScaler
sc_x=StandardScaler()
x=sc_x.fit_transform(x)
print(x)
```

```python
# Train Test Split

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
print(x.shape)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Ada Boost Model

'''Prajwal, why used Decision Tree becouse we know tat one most of the algorithms the logistic regression 
this dataset every thing coming as zero'''

model=DecisionTreeClassifier(criterion='entropy',max_depth=1)
AdaBoost=AdaBoostClassifier(estimator=model,n_estimators=10,learning_rate=1)
boostmodel=AdaBoost.fit(x_train,y_train)
display(boostmodel)
```

```python
# Prediction

y_pred=boostmodel.predict(x_test)
print(y_pred)
```

```python
# Accuracy

print('Accuracy Score:')
print(accuracy_score(y_test,y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```

```python
# Create Decision Tree Model

classifier=DecisionTreeClassifier(criterion='entropy')
classifier.fit(x_train,y_train)
display(classifier) 
```

```python
# Predictions

y_pred=classifier.predict(x_test)
print(y_pred)
```

```python
# Acuuracy

print('Accuracy Score:')
print(accuracy_score(y_test,y_pred))
print('Confusion Matrix:')
print(confusion_matrix(y_test,y_pred))
```
